```python
import yfinance as yf
import pandas as pd

print("⬇️ Downloading S&P500 data...")
spy = yf.download('^GSPC', start='2019-01-01', end='2024-12-01')

print(f"✅ Success! Shape: {spy.shape}")
print(f"Date range: {spy.index.min().date()} to {spy.index.max().date()}")
spy.head()
```

```python
# Save raw data
spy.to_csv('../02_Data/Raw/sp500_5yr.csv')
print("💾 Saved to: ../02_Data/Raw/sp500_5yr.csv")
```
