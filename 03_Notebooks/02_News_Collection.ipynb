```python
# Load and check
df = pd.read_csv('../02_Data/Raw/historical_news.csv')
df['date'] = pd.to_datetime(df['date'])

print(f"Total headlines: {len(df)}")
print(f"Date range: {df['date'].min().date()} to {df['date'].max().date()}")
print(f"\nFirst 5 headlines:")
for _, row in df.head().iterrows():
    print(f"{row['date'].date()}: {row['headline'][:60]}...")

print(f"\nLast 5 headlines:")
for _, row in df.tail().iterrows():
    print(f"{row['date'].date()}: {row['headline'][:60]}...")
```

```python
import matplotlib.pyplot as plt

# Count headlines per year
yearly_counts = df['date'].dt.year.value_counts().sort_index()

plt.figure(figsize=(12, 6))
yearly_counts.plot(kind='bar')
plt.title('News Headlines by Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Headlines', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../08_Visualization/news_timeline.png')
plt.show()

print("Timeline saved to 08_Visualization/news_timeline.png")
```
